```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
plt.style.use('fivethirtyeight')
plt.style.use('seaborn-white')
# plt.style.use('seaborn-paper')

import matplotlib.pylab as pylab
params = {'legend.fontsize': 'small',
         'axes.labelsize': 'medium',
         'axes.titlesize':'medium',
         'axes.titlesize':'large',
         'xtick.labelsize':'small',
         'ytick.labelsize':'small'}
pylab.rcParams.update(params)
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'
```

### Load labs measurments

- Velocities

```python
Vp_25_CAS = pd.read_csv('data/CAS/25/Vp.dat', sep="  ", names=['Pressure (MPa)','Vp_25_CAS'], header=None)
Vs_25_CAS = pd.read_csv('data/CAS/25/Vs.dat', sep="  ", names=['Pressure (MPa)','Vs_25_CAS'], header=None)
Vp_35_CAS = pd.read_csv('data/CAS/35/Vp.dat', sep="  ", names=['Pressure (MPa)','Vp_35_CAS'], header=None)
Vs_35_CAS = pd.read_csv('data/CAS/35/Vs.dat', sep="  ", names=['Pressure (MPa)','Vs_35_CAS'], header=None)

Vp_25_CH = pd.read_csv('data/CH/25/Vp.dat', sep="  ", names=['Pressure (MPa)','Vp_25_CH'], header=None)
Vs_25_CH = pd.read_csv('data/CH/25/Vs.dat', sep="  ", names=['Pressure (MPa)','Vs_25_CH'], header=None)
Vp_35_CH = pd.read_csv('data/CH/35/Vp.dat', sep="  ", names=['Pressure (MPa)','Vp_35_CH'], header=None)
Vs_35_CH = pd.read_csv('data/CH/35/Vs.dat', sep="  ", names=['Pressure (MPa)','Vs_35_CH'], header=None)

# concatenate velocities
V_25_CAS = pd.merge(Vp_25_CAS, Vs_25_CAS,  on='Pressure (MPa)')
V_25_CH = pd.merge(Vp_25_CH, Vs_25_CH,  on='Pressure (MPa)')
V_25 = pd.merge(V_25_CAS, V_25_CH, on='Pressure (MPa)')

V_35_CAS = pd.merge(Vp_35_CAS, Vs_35_CAS,  on='Pressure (MPa)')
V_35_CH = pd.merge(Vp_35_CH, Vs_35_CH,  on='Pressure (MPa)')
V_35 = pd.merge(V_35_CAS, V_35_CH, on='Pressure (MPa)')
V = pd.merge(V_25, V_35, on='Pressure (MPa)')
# V_35 = pd.merge(Vp_35, Vs_35, on='Pressure (MPa)')
# concatenate amplitude

# pp_25 = pd.merge(CO2_25, H2O_25, on='Pressure (MPa)')
# pp_35 = pd.merge(CO2_35, H2O_35, on='Pressure (MPa)')
# pp_25 = pd.merge(pp_25, V_25, on='Pressure (MPa)')
# pp_35 = pd.merge(pp_35, V_35, on='Pressure (MPa)')
# pp_25
```

- amplitude

```python
V['Amplitude_p_25_CAS'] = pd.read_csv('data/CAS/25/amp_P.dat', sep="  ", names=['Amplitude_p_25_CAS'], header=None)
V['Amplitude_p_25_CAS'] = np.flipud(V['Amplitude_p_25_CAS'])
V['Amplitude_s_25_CAS'] = pd.read_csv('data/CAS/25/amp_S.dat', sep="  ", names=['Amplitude_s_25_CAS'], header=None)
V['Amplitude_s_25_CAS'] = np.flipud(V['Amplitude_s_25_CAS'])
V['Amplitude_p_35_CAS'] = pd.read_csv('data/CAS/35/amp_P.dat', sep="  ", names=['Amplitude_p_35_CAS'], header=None)
V['Amplitude_p_35_CAS'] = np.flipud(V['Amplitude_p_35_CAS'])
V['Amplitude_s_35_CAS'] = pd.read_csv('data/CAS/35/amp_S.dat', sep="  ", names=['Amplitude_s_35_CAS'], header=None)
V['Amplitude_s_35_CAS'] = np.flipud(V['Amplitude_s_35_CAS'])

V['Amplitude_p_25_CH'] = pd.read_csv('data/CH/25/amp_P.dat', sep="  ", names=['Amplitude_p_25_CH'], header=None)
V['Amplitude_p_25_CH'] = np.flipud(V['Amplitude_p_25_CH'])
V['Amplitude_s_25_CH'] = pd.read_csv('data/CH/25/amp_S.dat', sep="  ", names=['Amplitude_s_25_CH'], header=None)
V['Amplitude_s_25_CH'] = np.flipud(V['Amplitude_s_25_CH'])
V['Amplitude_p_35_CH'] = pd.read_csv('data/CH/35/amp_P.dat', sep="  ", names=['Amplitude_p_35_CH'], header=None)
V['Amplitude_p_35_CH'] = np.flipud(V['Amplitude_p_35_CH'])
V['Amplitude_s_35_CH'] = pd.read_csv('data/CH/35/amp_S.dat', sep="  ", names=['Amplitude_s_35_CH'], header=None)
V['Amplitude_s_35_CH'] = np.flipud(V['Amplitude_s_35_CH'])
```

- dry velocities

```python
Vp_dry_CAS = pd.read_csv('data/CAS/dry/Vp.dat', sep="  ",names=['Pressure (MPa)','Vp_dry_CAS'], header=None)
# V['Amplitude_p_25_CAS'] = np.flipud(V['Amplitude_p_25_CAS'])
Vs_dry_CAS = pd.read_csv('data/CAS/dry/Vs.dat', sep="  ", names=['Pressure (MPa)','Vs_dry_CAS'], header=None)
# V['Amplitude_s_25_CAS'] = np.flipud(V['Amplitude_s_25_CAS'])
V_dry_CAS = pd.merge(Vp_dry_CAS, Vs_dry_CAS,  on='Pressure (MPa)')

Vp_dry_CH = pd.read_csv('data/CH/dry/Vp.dat', sep="  ", names=['Pressure (MPa)','Vp_dry_CH'], header=None)
# V['Amplitude_p_35_CAS'] = np.flipud(V['Amplitude_p_35_CAS'])
Vs_dry_CH = pd.read_csv('data/CH/dry/Vs.dat', sep="  ", names=['Pressure (MPa)','Vs_dry_CH'], header=None)
# V['Amplitude_s_35_CAS'] = np.flipud(V['Amplitude_s_35_CAS'])
V_dry_CH = pd.merge(Vp_dry_CH, Vs_dry_CH,  on='Pressure (MPa)')

V_dry = pd.merge(V_dry_CAS, V_dry_CH,  on='Pressure (MPa)')

```

```python
V = pd.merge(V, V_dry, on='Pressure (MPa)')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(15,10))

# plt.subplot(1,2,1)

ax1.plot(V['Pressure (MPa)'], V['Vp_25_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#D74E09', linewidth=2, label=r'Cairnside at 25ºC')
ax1.plot(V['Pressure (MPa)'], V['Vp_35_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#6E0E0A', linewidth=2, label=r'Cairnside at 35ºC')
ax1.plot(V['Pressure (MPa)'], V['Vp_25_CH'], marker='o', markeredgecolor='#125E8A', color='#197BBD', linewidth=2, label=r'Covey Hill at 25ºC')
ax1.plot(V['Pressure (MPa)'], V['Vp_35_CH'], marker='o', markeredgecolor='#125E8A', color='#125E8A', linewidth=2, label=r'Covey Hill at 35ºC')

ax1.plot(V['Pressure (MPa)'], V['Vp_dry_CAS'], ls='--', markeredgecolor='#6E0E0A', color='#D74E09', linewidth=2)
ax1.plot(V['Pressure (MPa)'], V['Vp_dry_CH'], ls='--', markeredgecolor='#125E8A', color='#197BBD', linewidth=2)

props = dict(boxstyle='round', facecolor='white', edgecolor='0.7', alpha=1)
ax1.text(0.04, 0.95, r'gaseous $\mathregular{CO_2}$', transform=ax1.transAxes, fontsize=12,verticalalignment='top', bbox=props)

ax1.yaxis.grid(True)
ax1.set_xlim([0,27])
ax1.set_ylim([3000,5000])
# ax1.set_xlabel('Pore Pressure (MPa)')
ax1.set_ylabel('Velocity (m/s)')
ax1.set_title('P-wave')
ax1.axvspan(0, 7, color='#A8A8A8', alpha=0.5, lw=0)

#Get artists and labels for legend and chose which ones to display
handles, labels = ax1.get_legend_handles_labels()
display = (0,1,2,3)

#Create custom artists
simArtist = plt.Line2D((0,1),(0,0), color='k', linestyle='--', alpha=0.7, linewidth=1.5)
# anyArtist = plt.Line2D((0,1),(0,0), color='k')

#Create legend from custom artist/label lists
legend = ax1.legend([handle for i,handle in enumerate(handles) if i in display]+[simArtist],
          [label for i,label in enumerate(labels) if i in display]+['Dry measurements'],loc='lower right', frameon=True, fancybox=True )
frame = legend.get_frame()
frame.set_facecolor('white')


# plt.subplot(1,2,2)
ax2.plot(V['Pressure (MPa)'], V['Vs_25_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#D74E09', linewidth=2)
ax2.plot(V['Pressure (MPa)'], V['Vs_35_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#6E0E0A', linewidth=2)
ax2.plot(V['Pressure (MPa)'], V['Vs_25_CH'], marker='o', markeredgecolor='#125E8A', color='#197BBD', linewidth=2)
ax2.plot(V['Pressure (MPa)'], V['Vs_35_CH'], marker='o', markeredgecolor='#125E8A', color='#125E8A', linewidth=2)

ax2.plot(V['Pressure (MPa)'], V['Vs_dry_CAS'], ls='--', markeredgecolor='#6E0E0A', color='#D74E09', linewidth=2)
ax2.plot(V['Pressure (MPa)'], V['Vs_dry_CH'], ls='--', markeredgecolor='#125E8A', color='#197BBD', linewidth=2)

ax2.text(0.04, 0.95, r'gaseous $\mathregular{CO_2}$', transform=ax2.transAxes, fontsize=12,verticalalignment='top', bbox=props)

ax2.yaxis.grid(True)
ax2.set_xlim([0,27])
ax2.set_ylim([2200, 3600])


ax2.set_title('S-wave')
ax2.axvspan(0, 7, color='#A8A8A8', alpha=0.5, lw=0)

ax3.plot(V['Pressure (MPa)'], V['Amplitude_p_25_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#D74E09', linewidth=2, label=r'Cairnside at 25ºC')
ax3.plot(V['Pressure (MPa)'], V['Amplitude_p_35_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#6E0E0A', linewidth=2, label=r'Cairnside at 35ºC')
ax3.plot(V['Pressure (MPa)'], V['Amplitude_p_25_CH'], marker='o', markeredgecolor='#125E8A', color='#197BBD', linewidth=2, label=r'Covey Hill at 25ºC')
ax3.plot(V['Pressure (MPa)'], V['Amplitude_p_35_CH'], marker='o', markeredgecolor='#125E8A', color='#125E8A', linewidth=2, label=r'Covey Hill at 35ºC')
ax3.text(0.04, 0.95, r'gaseous $\mathregular{CO_2}$', transform=ax3.transAxes, fontsize=12,verticalalignment='top', bbox=props)

ax3.set_ylabel('Amplitude')
ax3.set_xlim([0,27])
ax3.set_ylim([0, 1])
ax3.axvspan(0, 7, color='#A8A8A8', alpha=0.5, lw=0)
ax3.yaxis.grid(True)
ax3.set_xlabel('Pore Pressure (MPa)')


legend = ax3.legend(loc='upper right', frameon=True, fancybox=True)
frame = legend.get_frame()
frame.set_facecolor('white')

# ax2

ax4.plot(V['Pressure (MPa)'], V['Amplitude_s_25_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#D74E09', linewidth=2)
ax4.plot(V['Pressure (MPa)'], V['Amplitude_s_35_CAS'], marker='o', markeredgecolor='#6E0E0A', color='#6E0E0A', linewidth=2)
ax4.plot(V['Pressure (MPa)'], V['Amplitude_s_25_CH'], marker='o', markeredgecolor='#125E8A', color='#197BBD', linewidth=2)
ax4.plot(V['Pressure (MPa)'], V['Amplitude_s_35_CH'], marker='o', markeredgecolor='#125E8A', color='#125E8A', linewidth=2)
ax4.text(0.04, 0.95, r'gaseous $\mathregular{CO_2}$', transform=ax4.transAxes, fontsize=12,verticalalignment='top', bbox=props)

ax4.set_xlim([0,27])
ax4.set_ylim([0, 1])
ax4.axvspan(0, 7, color='#A8A8A8', alpha=0.5, lw=0)
ax4.yaxis.grid(True)
ax4.set_xlabel('Pore Pressure (MPa)')


plt.savefig("fig/results_lab.pdf", bbox_inches='tight')

```
